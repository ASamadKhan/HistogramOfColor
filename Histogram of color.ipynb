```python
import numpy as np
import matplotlib.image as mpimage
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading an image
image = mpimage.imread("cutout1.jpg")
#displaing an image
plt.imshow(image)
```

### to get orderly bin sizes use  bin=32 and range (0,256)

```python
rhist = np.histogram(image[:,:,0],bins=32,range=(0,256))
ghist = np.histogram(image[:,:,1],bins=32, range=(0,256))
bhist = np.histogram(image[:,:,2],bins=32, range=(0,256))
rhist
```

# np.histogram
np.histogram() function will return a tuple of two arrays,e.g., in rhist case<br>
1. rhist[0] contains the count in each of the bin <br>
2. rhist[1] contains the bin edges (so it is one element longer than rhist[0]) <br>
we will compute the bin_centers from bing edges. Each of the histograms in this case have the same bins, so I'll just use the rhist bin edges:

```python
bin_edges = rhist[1]
bin_centers =(bin_edges[1:]+bin_edges[0:len(bin_edges)-1])/2
```

# Displaying histograms

```python
fig=plt.figure(figsize=(30,5))
plt.subplot(131)
plt.bar(bin_centers,rhist[0])
plt.xlim(0,256)
plt.title("Red color Histogram")
plt.subplot(132)
plt.bar(bin_centers,ghist[0])
plt.xlim(0,256)
plt.title("Green color Histogram")
plt.subplot(133)
plt.bar(bin_centers,bhist[0])
plt.xlim(0,256)
plt.title("Blue color Histogram")
```

# Histogram feature vector
concatenation will give us the histogram feature vector 

```python
hist_feature = np.concatenate((rhist[0],ghist[0],bhist[0]))
```
